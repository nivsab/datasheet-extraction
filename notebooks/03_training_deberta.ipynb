```python
"""
create_val_split.py
===================
יוצר קובץ val.json מתוך הנתונים המעובדים.
מריצים פעם אחת על המחשב המקומי לפני העלאה ל-Colab.
"""

import os
import json
import glob
import random

# ============================================================
# הגדרות
# ============================================================
INPUT_DIR  = r"C:\Users\nivsa\Generation of Synthetic Training Data\embedded\data\processed"
OUTPUT_DIR = r"C:\Users\nivsa\Generation of Synthetic Training Data\embedded\data\splits"
VAL_SPLIT  = 0.1   # 10% לvalidation
SEED       = 42

# ============================================================
# טעינת כל הנתונים
# ============================================================
random.seed(SEED)

print("טוען נתונים...")
all_samples = []
json_files = glob.glob(os.path.join(INPUT_DIR, "**", "*.json"), recursive=True)

if not json_files:
    raise FileNotFoundError(f"לא נמצאו קבצי JSON ב: {INPUT_DIR}")

for path in json_files:
    try:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        if isinstance(data, list):
            all_samples.extend(data)
        else:
            all_samples.append(data)
    except Exception as e:
        print(f"  דילוג על {os.path.basename(path)}: {e}")

print(f"נטענו {len(all_samples)} samples מתוך {len(json_files)} קבצים.")

# ============================================================
# פיצול
# ============================================================
random.shuffle(all_samples)
val_size   = max(1, int(len(all_samples) * VAL_SPLIT))
val_samples = all_samples[:val_size]

print(f"Val set: {len(val_samples)} samples ({VAL_SPLIT*100:.0f}%)")

# ============================================================
# שמירה
# ============================================================
os.makedirs(OUTPUT_DIR, exist_ok=True)
val_path = os.path.join(OUTPUT_DIR, "val.json")

with open(val_path, "w", encoding="utf-8") as f:
    json.dump(val_samples, f, ensure_ascii=False, indent=2)

print(f"נשמר: {val_path}")
print(f"גודל הקובץ: {os.path.getsize(val_path) / 1024 / 1024:.1f} MB")
```
